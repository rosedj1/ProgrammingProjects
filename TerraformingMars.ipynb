```python
## Slice inputDF based on value of row_label
## Make 
def makeDF(inputDF, row_label, value, showInfo=True):
    col_list = []
    
    ## Create new, empty DF
    ## Make sure it has total number of indices (rows)
    outputDF = pd.DataFrame(data=None, index=df.index, columns=[])
    
    if showInfo == True: print("Starting output DataFrame and all information:", "\n", outputDF)
    if showInfo == True: print("\n", "column labels:", inputDF.columns)
    
    for col in inputDF.columns:
        ## Make list of columns which we want to keep ##
        if inputDF[col].loc[row_label] == value: col_list.append(col)
    print(col_list)
```
